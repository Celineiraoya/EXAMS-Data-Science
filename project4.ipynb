```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import plotly.express as px
```

```python
#What is the distribution of prices among different subcategories?
num_records = 1000
Sub_Category = ['Snacks', 'Bakery $ Desserts', 'Snacks', 'Bakery $ Desserts' ]
prices = ['$21.99','$17.49', '$22.99', '$49.99']
sub_category_data = np.random.choice(Sub_Category, size=num_records)
product_data = pd.DataFrame(data={'Sub_Category': Sub_Category, 'prices': prices})
print(product_data.head())
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(x='Sub_Category', y='prices', data = product_data, palette='summer')
plt.title('PRICE DISTRIBUTED AMONG DIFFERENT SUBCATEGORIES')
plt.xlabel('Sub_Category', color = 'Blue', fontsize = 20)
plt.ylabel('Prices', color = 'Blue', fontsize = 20)
plt.show()
```

```python
#How does the discount percentage vary across various product categories?
np.random.seed(42)
num_records=1000
Sub_Category = np.random.choice (['Snacks', 'Bakery $ Desserts', 'Snacks', 'Bakery $ Desserts' ], size=num_records)
Discount = np.random.uniform(0,10, size=num_records)
product_data = pd.DataFrame(data={'Sub_Category': Sub_Category, 'Discount': Discount})
print(product_data.head())
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='Sub_Category', y='Discount', data=product_data, color = 'Blue')
plt.title('Discount accross different Subcategories', color = 'Black')
plt.xlabel('Sub category', color = 'crimson', fontsize=20)
plt.ylabel('Discount', color = 'crimson', fontsize = 20)
plt.show()
```

```python
#Is there any correlation between product ratings and prices?
num_records= 1000
Prices = np.random.uniform(20, 200, size=num_records)
Ratings = np.random.uniform(1, 5, size=num_records)
product_data = pd.DataFrame(data={'Prices': Prices, 'Rating': Ratings})
print(product_data.head())
```

```python
plt.figure(figsize=(12,10))
sns.swarmplot(x='Prices', y='Rating', data = product_data, color = 'skyblue')
plt.title('Correlation', color = 'Blue', fontsize=20)
plt.xlabel('Prices')
plt.ylabel('Ratings')
plt.show()
```

```python
#What are the most common currencies used in the dataset?
data = {'currency': ['USD']}
dataset = pd.DataFrame(data)
print(dataset.head())
```

```python
#How do product features differ across subcategories?
np.random.seed(40)
num_records=1000
Sub_Category = np.random.choice (['Snacks', 'Bakery $ Desserts', 'Snacks', 'Bakery $ Desserts' ], size=num_records)
features = np.random.uniform(0,1, size=num_records)
product_data = pd.DataFrame(data={'Sub_Category': Sub_Category, 'features': features})
print(product_data.head())
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Sub_Category', y='features', data=product_data, color='Blue')
plt.title('Features accross Sub_Categories', color = 'black', fontsize=20)
plt.xlabel('Sub_Category', color = 'Blue', fontsize=20)
plt.ylabel('Feature Value', color = 'Brown', fontsize=20)
plt.show()
```

```python
#What is the overall distribution of product ratings?
num_records = 1000
Ratings = np.random.uniform(1,5, size=num_records)
product_data = pd.DataFrame(data={'Rating': Ratings})
print(product_data.head())
```

```python
plt.figure(figsize=(10,6))
sns.histplot(product_data['Rating'], bins=20, kde=True, color = 'crimson')
plt.title('Ratings of Products', color = 'Blue', fontsize = 15)
plt.xlabel('Ratings', color = 'Blue', fontsize = 20)
plt.ylabel('Frequency', color = 'Blue', fontsize = 20)
plt.show()
```

```python
#Are there any outliers in the pricing of products with high discounts?
num_records = 1000
Price =  np.random.uniform(50, 500, size = num_records)
Discount = np.random.uniform(10, 50, size = num_records)
product_data = pd.DataFrame(data={'price': Price, 'Discount': Discount})
print(product_data.head())
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='price', y='Discount', data=product_data, color = 'Blue')
plt.title('Outliers')
plt.xlabel('Price')
plt.ylabel('Discount')

plt.show()
```

```python
#How do different currencies compare when considering average product prices?
num_records = 1000
Currency = ['USD', 'EUR', 'CAD', 'JPY', 'GBP']
Price =  np.random.uniform(10, 100, size = num_records)
product_data = pd.DataFrame(data={'Currency': np.random.choice(Currency, size=num_records), 'Price': Price})
print(product_data.head())
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Currency', y='Price', data=product_data, color = 'crimson')
plt.title('Distribution pf Product Prices', color = 'Blue', fontsize = 15)
plt.xlabel('Currency', color = 'Blue', fontsize = 20)
plt.ylabel('Product Price', color = 'Blue', fontsize = 20)
plt.show()
```

```python
#Is there a relationship between the length of product descriptions and their ratings?
num_records = 1000
Ratings = np.random.uniform(1,5, size=num_records)
Product_Description = np.random.uniform(50, 500, size = num_records)
product_data = pd. DataFrame(data={'Rating': Ratings, 'Product_Description': Product_Description })
print(product_data.head())
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Product_Description', y = 'Rating', data=product_data, color = 'Black')
plt.title('Description vs Ratings')
plt.xlabel('Description')
plt.ylabel('Rating')
plt.show()
```
